```python
import json
import os
import numpy as np
from xml.etree.ElementTree import Element
from xml.etree.ElementTree import XMLParser
import xml.etree.ElementTree as ET

directory = r'C:\Projects\cvat-parser'
# Iterate over each task in the tasks of xml file
tasks= []
for filename in os.listdir(directory):
    # Check if the file is a JSON file
    if filename.endswith(".xml"):
        #print(filename)
        tree = ET.parse(filename)
        #print(tree)
        root = tree.getroot()
        #print(root)
        xml_str = ET.tostring(root).decode()
        #print(xml_str)
        for children in root:
            children_tag= children.tag
            #print(children_tag)
            if children_tag=='meta':
                #print(children)
                for child in children:
                    #print(child)
                    if child.tag == 'project':
                        for c in child:
                            #print(c)
                            if c.tag== 'tasks':
                                #print("tag: ", c.tag)
                                for gc in c:
                                    #print(ET.tostring(gc).decode())
                                    #print(gc)
                                    task_dict= {}
                                    for gc2 in gc:
                                        #print(gc2.tag)
                                        if gc2.tag== 'id':
                                            task_dict['id']= gc2.text
                                        elif gc2.tag== 'name':
                                            task_dict['name']= gc2.text
                                        elif gc2.tag== 'owner':
                                            # task_dict['name']= gc2.text
                                            for gc3 in gc2:
                                                #print(gc3.tag)
                                                if gc3.tag=='username':
                                                   task_dict['owner_name']= gc3.text
                                                if gc3.tag=='email':
                                                   task_dict['owner_email']= gc3.text
                                        elif gc2.tag== 'assignee':
                                            # task_dict['name']= gc2.text
                                            for gc4 in gc2:
                                                #print(gc3.tag)
                                                if gc4.tag=='username':
                                                   task_dict['assignee_name']= gc4.text
                                                if gc4.tag=='email':
                                                   task_dict['assignee_email']= gc4.text
                                    #print(task_dict)
                                    tasks.append(task_dict)
        print("tasks info list: ")
        print(tasks)
        task_id_wise_box_count= {}
        for children in root:
            children_tag= children.tag
            #print(children_tag)
            if children.tag=='image':
                #print(children.attrib['task_id'])
                if children.attrib['task_id'] not in task_id_wise_box_count.keys():
                    #id_wise_box_count[children.attrib['task_id']]= 1
                    count= 0
                    for c in children:
                        count= count+1
                    task_id_wise_box_count[children.attrib['task_id']]= count
                else:
                    count= 0
                    for c in children:
                        count= count+1
                    task_id_wise_box_count[children.attrib['task_id']]= task_id_wise_box_count[children.attrib['task_id']]+count
        print("task wise object count: ")
        print(task_id_wise_box_count)
        annotator = {}
        for task in tasks:
            #print(task['id'], task['owner_name'])
            if task['owner_name'] not in annotator.keys():
                annotator[task['owner_name']]= task_id_wise_box_count[task['id']]
            else:
                annotator[task['owner_name']]= annotator[task['owner_name']]+ task_id_wise_box_count[task['id']]
        print("object count: ")
        print(annotator)
                
                
                            
```
